## Basic Plots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
x = np.linspace(1,10,10)
y = np.log10(x)
plt.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.plot(x, y+0.5, 'ro--')  # shortcut!
```

There are infinite options, here's some basics.

**Colors**
* b : blue
* g: green
* r : red
* c : cyan
* m : magenta
* y : yellow
* k : black
* w : white

**Markers**
* . small point
* o bigger point
* s square
* x x


**Line Styles**
* \- solid
* : dotted
* -- dashed

```python
x = np.linspace(0, 10, 25)
y1 = np.cos(x)
offset = 0.5
y2 = y1 + offset
y3 = y2 + offset
y4 = y3 + offset

plt.plot(x,y1,'b-')   # blue line
plt.plot(x,y2,'r:')   # red dotted line
plt.plot(x,y3,'gs')   # green squares (no line)
plt.plot(x,y4,'mo--') # magenta points and a dashed line
plt.show()
```

# Making things pretty

```python
plt.figure(figsize=(8,5))  # set figure size in inches

plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')

plt.xlim([0,10])  # Set axis ranges
plt.ylim([-2,2])

plt.xlabel('Time [s]')
plt.ylabel('Distance [m]')
plt.title('A Properly Labelled Plot')
plt.grid()
plt.legend();

plt.show()
```

## Scatter Plots
'scatter' is similar to 'plot' except every point can be rendered differently

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar();  # show color scale
```

## Histograms

```python
data = np.random.randn(1000)
plt.hist(data)
plt.show()
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

## Subplots
Usage: `subplot(nrows, ncols, index, **kwargs)`


```python
x = np.linspace(0,10,50)

plt.figure(figsize=(12,4))  # figure size in inches

plt.subplot(1,3,1)  # 1 row, 3 columns, plot 1
plt.plot(x,np.sin(x),'r-')

plt.subplot(1,3,2)  # plot 2
plt.plot(x,np.exp(-x),'g.:')

plt.subplot(1,3,3)  # plot 3
plt.plot(x,np.floor(x),'b.:')

plt.tight_layout()
plt.show()
```
